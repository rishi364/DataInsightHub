```python

import pandas as pd 
df = pd.read_csv('HR_Analytics.csv')
```

```python

print(df.isnull().sum())
```

```python

df = df.dropna()

df = df.fillna({
    'DailyRate' : df['DailyRate'].median(),
    'DitanceFromHome' : df['DistanceFromHome'].median(),
    'MonthlyIncome' : df['MonthlyIncome'].median(),
})
```

```python
df = df.drop_duplicates()
```

```python
df.to_csv('HR_Analytics.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cleaned_dataset.csv')

age_filtered_df = df[(df['AgeGroup'] == '18-25')]

plt.figure(figsize=(10, 6))
age_filtered_df['Department'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Department Distribution for Employees Aged 18-25')
plt.xlabel('Department')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
age_filtered_df['EducationField'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Education Field Distribution for Employees Aged 18-25')
plt.xlabel('Education Field')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
education_job_satisfaction = df.groupby('EducationField')['JobSatisfaction'].mean()

plt.figure(figsize=(10, 6))
education_job_satisfaction.plot(kind='bar', color='orange')
plt.title('Average Job Satisfaction by Education Field')
plt.xlabel('Education Field')
plt.ylabel('Average Job Satisfaction')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
salary_hike_distribution = df['PercentSalaryHike'].value_counts()

plt.figure(figsize=(8, 8))
salary_hike_distribution.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired(range(len(salary_hike_distribution))))
plt.title('Percentage Salary Hike Distribution')
plt.ylabel('')
plt.grid(True)
plt.show()

```
